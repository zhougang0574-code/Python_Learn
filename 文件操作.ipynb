```python

from debugpy.common.json import JsonEncoder


# 写一个函数，读取 JSON 文件，要求处理以下两种异常：
# 1. 文件不存在，打印 "文件不存在"
# 2. JSON 格式错误，打印 "JSON格式错误"
# 正常读取成功则打印文件内容

# 测试用例：
# read_json("users.json")       # 正常文件
# read_json("no_file.json")     # 不存在的文件
# read_json("broken.json")      # 内容写入 "this is not json" 模拟格式错误

import json
from json import JSONDecodeError

def do_except(file_path:str):
    try:
        with open(file_path,"r",encoding="utf-8") as f:
            message = json.load(f)
            print(message)
    except FileNotFoundError as e:
        print (f"文件不存在")
    except JSONDecodeError as e:
        print (f"JSON格式错误")

do_except("test.txt")
```

```python
# 有一个 JSON 字符串（模拟接口返回的数据），解析后把其中 status 为 "success" 的记录
# 提取出 id 和 message 写入 result.json 文件
# s = '{"code": 200, "data": [{"id": 1, "status": "success", "message": "订单创建成功"}, {"id": 2, "status": "fail", "message": "库存不足"}, {"id": 3, "status": "success", "message": "支付成功"}]}'
# 期望 result.json 内容：
# [
#   {"id": 1, "message": "订单创建成功"},
#   {"id": 3, "message": "支付成功"}
# ]

s = '{"code": 200, "data": [{"id": 1, "status": "success", "message": "订单创建成功"}, {"id": 2, "status": "fail", "message": "库存不足"}, {"id": 3, "status": "success", "message": "支付成功"}]}'

import json

js = json.loads(s)
data = js["data"]
success_data = [ m for m in data if m["status"] == "success"]
print(success_data)
with open("result.json","w",encoding="utf-8") as w:
    json.dump(success_data,w,ensure_ascii=False,indent=2)
```

```python
# 读取 users.json 文件，把 age 大于 25 的用户筛选出来
# 把筛选后的结果写入新文件 filtered.json
# 然后打印 filtered.json 的内容

import json

with open("user.json","r",encoding="utf-8") as r ,open("filtered.json","w",encoding="utf-8") as w:
    data = json.load(r)
    filtered = [user for user in data if user["age"] >25]
    json.dump(filtered,w,ensure_ascii=False,indent=2)
    json.dumps()
    print(filtered)

```

```python
# 遍历当前目录下所有 .txt 文件，打印每个文件的文件名和文件大小
# 例如输出：
# test.txt - 123 bytes
# output.txt - 456 bytes

# 提示：文件大小用 p.stat().st_size 获取，单位是字节
from pathlib import Path

path = Path(".")
for f in path.glob("*.txt"):
    print(f"{f.name} - {f.stat().st_size}")
```

```python
# 读取 test.txt 文件，统计文件中每个单词出现的次数，打印出来
# 例如 test.txt 内容：
# hello world
# hello python
# world world
# 期望输出：
# {'hello': 2, 'world': 3, 'python': 1}
with open("test.txt",'r') as f:
    kv = {}
    for line in f:
        ls = line.strip().split(" ")
        for subkv in ls:
            if subkv in kv:
                kv[subkv] += 1
            else:
                kv.update({subkv:1})
    print(kv)
```

```python
# 读取 test.txt 文件，把每一行的内容反转后写入新文件 output.txt
# 例如 test.txt 内容：
# Hello
# World
# Python
# 期望 output.txt 内容：
# olleH
# dlroW
# nohtyP

with open("test.txt",'r') as f,open("output.txt",'a') as w:
    for line in f:
        obl =line.strip()[::-1]
        w.write(obl+"\n")
```

```python
# 读取一个 txt 文件，统计文件中一共有多少行，打印出来
# 例如 test.txt 内容是：
# Hello
# World
# Python
# 期望输出：3

def read_lines(file_path:str):
    line = 0
    with open(file_path,"r",encoding="utf-8") as f:
        # 全部读取
        content = f.readlines()
        line = len(content)
    return line

read_lines("test.txt")
```

```python
# 在当前目录创建一个 test.txt 文件
# 写入以下三行内容：
# Hello
# World
# Python
# 然后读取文件内容并打印出来

with open("test.txt","w",encoding="utf-8") as f :
    f.write("Hello\n")
    f.write("World\n")
    f.write("Python\n")

with open("test.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```
