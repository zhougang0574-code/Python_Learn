### 5、练习题

```python
# 遍历当前目录下所有子目录和文件
# 打印出来，目录用 [DIR] 标记，文件用 [FILE] 标记
# 例如输出：
# [DIR]  logs
# [FILE] test.txt
# [FILE] output.txt

from pathlib import Path

path = Path(".")
for file in path.iterdir():
    print(("[DIR]" if file.is_dir() else "[FILE]") + file.name)
```

```python
# 在当前目录创建一个 logs 目录，目录已存在不报错
# 在 logs 目录下创建 app.log 文件，写入以下内容：
# 2024-01-01 启动成功
# 2024-01-02 运行中
# 2024-01-03 关闭
# 然后读取并打印 app.log 的内容

from pathlib import Path

path = Path("logs").mkdir(exist_ok=True)
with open("logs/app.log","w") as f:
    f.write("2024-01-01 启动成功\n")
    f.write("2024-01-02 运行中\n")
    f.write("2024-01-03 关闭\n")

with open("logs/app.log","r") as f:
    content = f.read()
    print(content)
```

```python
# 遍历当前目录下所有 .txt 文件，打印每个文件的文件名和文件大小
# 例如输出：
# test.txt - 123 bytes
# output.txt - 456 bytes

# 提示：文件大小用 p.stat().st_size 获取，单位是字节
from pathlib import Path

path = Path(".")
for f in path.glob("*.txt"):
    print(f"{f.name} - {f.stat().st_size}")
```

```python
# 读取 test.txt 文件，统计文件中每个单词出现的次数，打印出来
# 例如 test.txt 内容：
# hello world
# hello python
# world world
# 期望输出：
# {'hello': 2, 'world': 3, 'python': 1}
with open("test.txt",'r') as f:
    kv = {}
    for line in f:
        ls = line.strip().split(" ")
        for subkv in ls:
            if subkv in kv:
                kv[subkv] += 1
            else:
                kv.update({subkv:1})
    print(kv)
```

```python
# 读取 test.txt 文件，把每一行的内容反转后写入新文件 output.txt
# 例如 test.txt 内容：
# Hello
# World
# Python
# 期望 output.txt 内容：
# olleH
# dlroW
# nohtyP

with open("test.txt",'r') as f,open("output.txt",'a') as w:
    for line in f:
        obl =line.strip()[::-1]
        w.write(obl+"\n")
```

```python
# 读取一个 txt 文件，统计文件中一共有多少行，打印出来
# 例如 test.txt 内容是：
# Hello
# World
# Python
# 期望输出：3

def read_lines(file_path:str):
    line = 0
    with open(file_path,"r",encoding="utf-8") as f:
        # 全部读取
        content = f.readlines()
        line = len(content)
    return line

read_lines("test.txt")
```

```python
# 在当前目录创建一个 test.txt 文件
# 写入以下三行内容：
# Hello
# World
# Python
# 然后读取文件内容并打印出来

with open("test.txt","w",encoding="utf-8") as f :
    f.write("Hello\n")
    f.write("World\n")
    f.write("Python\n")

with open("test.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```

### 6、JSON文件

```python
# 读取json

import json

with open("config.json","r",encoding="utf-8") as f:
    data = json.load(f)
    print(data)

# 直接读取json格式的字符串
data2 = '{"name": "Alice", "age": 30}'
result = json.loads(data2)
print(result)
```

```python
# 写入JSON
import json

data = {"name": "Alice", "age": 30}

with open("config.json","w",encoding="utf-8") as f:
    json.dump(data,f,ensure_ascii=False,indent=2)
```
