### 1、成绩等级判断

```python
# 写一个判断：输入分数 score = 73
# >=90 是 "A"，>=80 是 "B"，>=60 是 "C"，否则是 "D"，打印等级
# 期望输出：C

# 在这里写你的代码

def func (score:float):
    if score >= 90:
        print("A")
    elif score >= 80:
        print("B")
    elif score >= 60:
        print("C")
    else:
        print("D")

func(73)
```

### 2、三元表达式

```python
# 有一个列表 nums = [3, -1, 0, 8, -5]
# 用一行三元表达式判断每个数是 "正数"、"负数" 还是 "零"，存到新列表里并打印
# 期望输出：['正数', '负数', '零', '正数', '负数']

# 在这里写你的代码

nums = [3, -1, 0, 8, -5]
new_nums = [ "正数" if i > 0 else ("零" if i == 0 else "负数")  for i in nums ]
print(new_nums)
```

### 3、for 循环 + enumerate

```python
# 有一个列表 fruits = ["apple", "banana", "cherry"]
# 用 enumerate 打印每个水果的下标和名字，格式：下标: 名字
# 期望输出：
# 0: apple
# 1: banana
# 2: cherry

# 在这里写你的代码

fruits = ["apple", "banana", "cherry"]

for k,i in enumerate(fruits):
    print(f"{k}: {i}")
```

### 4、while + break 猜数字

```python
# 有一个目标数字 target = 7，和一个待猜列表 guesses = [3, 9, 7, 5]
# 用 while 循环依次检查 guesses 里的每个数字（用下标遍历），
# 猜中 target 就打印 "猜中了: 7" 并停止循环；如果列表猜完了还没猜中，打印 "没猜中"
# 期望输出：猜中了: 7

# 在这里写你的代码

def func(target:int,guess:list):
    i = 0
    while i < len(guess):
        if target == guess[i]:
            print(f"猜中了：{target}")
            break
        i += 1
    else:
        print("没猜中")

func(7,[3,9,7,5])
```

### 5、for-else 查找

```python
# 有一个列表 nums = [2, 4, 6, 8]，和一个目标 target = 5
# 用 for-else 遍历 nums 查找 target：找到就打印 "找到了" 并 break，
# 循环正常结束（没找到）就在 else 块里打印 "没找到"
# 期望输出：没找到

# 在这里写你的代码


def func(target:int,guess:list):
    for i in range(len(guess)):
        if target == guess[i]:
            print("找到了")
            break
    else:
        print("没找到")

func(5,[2,4,6,8])
```
