# 控制流 · 知识点

### 1、if / elif / else 条件分支

```python
# Python 用缩进划分代码块，没有 {}，冒号 : 后面必须换行缩进
score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "D"
print(grade)

# 没有 switch/case（3.10 以前），多分支基本靠 if/elif/else 链
# 条件不需要括号，但写了也不报错：if (score >= 60): 也合法，只是不推荐
```

### 2、比较运算符 & 逻辑运算符

```python
# 比较运算符：== != > < >= <=，和 Java 一样
# 逻辑运算符是单词 and / or / not，不是 && || !
a, b = 5, 10
print(a < b and b < 20)
print(a > b or b > 5)
print(not a == b)

# and / or 是短路求值，会返回其中一个操作数本身，不一定是布尔值
print(0 or "default")     # 0 是假值，所以返回 "default"
print("hi" and "there")   # 第一个真值，继续判断第二个，返回 "there"

# == 比较的是值是否相等，is 比较的是是否是同一个对象（类似 Java 的 ==）
x = [1, 2]
y = [1, 2]
print(x == y)   # True，值相等
print(x is y)   # False，不是同一个对象

# 链式比较，Python 特有，等价于 1 < x and x < 10
x = 5
print(1 < x < 10)
```

### 3、三元表达式（条件表达式）

```python
# Python 没有 ? : 三元运算符，写法是 值A if 条件 else 值B
n = 7
result = "偶数" if n % 2 == 0 else "奇数"
print(result)

# 等价于 Java 的: String result = n % 2 == 0 ? "偶数" : "奇数";
# 可以嵌套，但嵌套多了很难读，不建议套两层以上
x = 5
label = "正" if x > 0 else ("零" if x == 0 else "负")
print(label)
```

### 4、for 循环

```python
# Python 的 for 是 for-each 风格，直接遍历元素，没有传统的 for(int i=0;...) 写法
for x in [1, 2, 3]:
    print(x)

# 需要下标时用 range()，range(start, stop, step)，左闭右开
for i in range(5):
    print(i, end=" ")
print()
for i in range(2, 10, 2):
    print(i, end=" ")
print()

# 同时要下标和值，用 enumerate()，比手写 i 计数器更地道
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(i, fruit)

# 同时遍历多个序列，用 zip()，按最短的那个截断
names = ["Tom", "Jerry"]
ages = [18, 20]
for name, age in zip(names, ages):
    print(name, age)
```

### 5、while 循环

```python
# while 在条件为真时反复执行，写法和 Java 基本一致
n = 3
while n > 0:
    print(n)
    n -= 1   # 没有 n--，Python 不支持 ++ / --，只能 n -= 1 或 n = n - 1

# while True 配合 break 实现"至少执行一次直到满足条件才退出"，
# 对应 Java 里没有内置 do-while 关键字时的常见替代写法
count = 0
while True:
    count += 1
    if count >= 3:
        break
```

### 6、break / continue 与 for-else、while-else

```python
# break 跳出整个循环，continue 跳过本次循环剩余部分，进入下一次迭代
for x in range(10):
    if x == 3:
        continue   # 跳过 3
    if x == 6:
        break      # 到 6 就停止
    print(x, end=" ")
print()

# for/while 后面可以接 else，Java 没有这个语法
# else 块在循环【正常结束（没有被 break 中断）】时执行，常用来表达"没找到"
target = 100
nums = [1, 2, 3, 4, 5]
for n in nums:
    if n == target:
        print("找到了")
        break
else:
    print("没找到")
```
