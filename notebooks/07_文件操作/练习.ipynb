### 1、JSONL 文件读写与统计

```python

# 把订单事件写入 orders.jsonl 文件，每行一个 JSON 对象，中文不要转义成 \uXXXX
# orders = [
#     {"order_id": 1001, "status": "已支付", "amount": 199.0},
#     {"order_id": 1002, "status": "已取消", "amount": 59.0},
#     {"order_id": 1003, "status": "已支付", "amount": 320.0},
# ]
# 写入后逐行读取 orders.jsonl，累加 status 为"已支付"的订单金额，打印总额
# 期望输出：519.0

# 在这里写你的代码

import json

orders = [
    {"order_id": 1001, "status": "已支付", "amount": 199.0},
    {"order_id": 1002, "status": "已取消", "amount": 59.0},
    {"order_id": 1003, "status": "已支付", "amount": 320.0},
]

with open("orders.jsonl","w",encoding="utf-8") as f:
    for order in orders:
        f.write(json.dumps(order,ensure_ascii=False) + "\n")

with open("orders.jsonl","r",encoding="utf-8") as f:
    count = 0
    for j in f:
        order = json.loads(j)
        count += order["amount"] if order["status"] == "已支付" else 0
    print(count)
```

### 2、CSV 文件读写 + 转 JSON

```python
# 1. 把员工信息写入 employees.csv，要求带表头（name, department, salary）
# employees = [
#     {"name": "张三", "department": "技术部", "salary": 15000},
#     {"name": "李四", "department": "销售部", "salary": 12000},
#     {"name": "王五", "department": "技术部", "salary": 18000},
# ]
# 2. 用 DictReader 读取 employees.csv，筛选出 department 为"技术部"的员工姓名，打印出来
#    期望输出：['张三', '王五']
#    注意：DictReader 读出来的字段都是字符串
# 3. 把全部员工数据转换写入 employees.json，要求 salary 字段是 int 类型（不能是字符串）

# 在这里写你的代码
import csv

employees = [
    {"name": "张三", "department": "技术部", "salary": 15000},
    {"name": "李四", "department": "销售部", "salary": 12000},
    {"name": "王五", "department": "技术部", "salary": 18000},
]

with open("employees.csv",'w',encoding="utf-8",newline="") as f:
    writer = csv.DictWriter(f,fieldnames=['name','department','salary'])
    for employee in employees:
        print(employee)
        writer.writerow(employee)

with open("employees.csv",'r',encoding="utf-8",newline="") as f:
    reader = csv.DictReader(f)  # 自动把第一行当表头，后面每行转成 {表头: 值} 的字典
    for row in reader:
        print(row)
```
