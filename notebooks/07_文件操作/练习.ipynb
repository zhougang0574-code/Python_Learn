```python
# 在当前目录创建一个 test.txt 文件
# 写入以下三行内容：
# Hello
# World
# Python
# 然后读取文件内容并打印出来

with open("test.txt","w",encoding="utf-8") as f :
    f.write("Hello\n")
    f.write("World\n")
    f.write("Python\n")

with open("test.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```

```python
# 读取一个 txt 文件，统计文件中一共有多少行，打印出来
# 例如 test.txt 内容是：
# Hello
# World
# Python
# 期望输出：3

def read_lines(file_path:str):
    line = 0
    with open(file_path,"r",encoding="utf-8") as f:
        # 全部读取
        content = f.readlines()
        line = len(content)
    return line

read_lines("test.txt")
```

```python
# 读取 test.txt 文件，把每一行的内容反转后写入新文件 output.txt
# 例如 test.txt 内容：
# Hello
# World
# Python
# 期望 output.txt 内容：
# olleH
# dlroW
# nohtyP

with open("test.txt",'r') as f,open("output.txt",'a') as w:
    for line in f:
        obl =line.strip()[::-1]
        w.write(obl+"\n")
```

```python
# 读取 test.txt 文件，统计文件中每个单词出现的次数，打印出来
# 例如 test.txt 内容：
# hello world
# hello python
# world world
# 期望输出：
# {'hello': 2, 'world': 3, 'python': 1}
with open("test.txt",'r') as f:
    kv = {}
    for line in f:
        ls = line.strip().split(" ")
        for subkv in ls:
            if subkv in kv:
                kv[subkv] += 1
            else:
                kv.update({subkv:1})
    print(kv)
```

```python
# 遍历当前目录下所有 .txt 文件，打印每个文件的文件名和文件大小
# 例如输出：
# test.txt - 123 bytes
# output.txt - 456 bytes

# 提示：文件大小用 p.stat().st_size 获取，单位是字节
from pathlib import Path

path = Path(".")
for f in path.glob("*.txt"):
    print(f"{f.name} - {f.stat().st_size}")
```

```python
# 写一个函数，读取 JSON 文件，要求处理以下两种异常：
# 1. 文件不存在，打印 "文件不存在"
# 2. JSON 格式错误，打印 "JSON格式错误"
# 正常读取成功则打印文件内容

# 测试用例：
# read_json("users.json")       # 正常文件
# read_json("no_file.json")     # 不存在的文件
# read_json("broken.json")      # 内容写入 "this is not json" 模拟格式错误

import json
from json import JSONDecodeError

def do_except(file_path:str):
    try:
        with open(file_path,"r",encoding="utf-8") as f:
            message = json.load(f)
            print(message)
    except FileNotFoundError as e:
        print (f"文件不存在")
    except JSONDecodeError as e:
        print (f"JSON格式错误")

do_except("test.txt")
```

```python
# 写一个函数，读取一个 JSON 文件
# 如果文件不存在，自动创建文件并写入空列表 []
# 如果文件存在，读取内容并返回
# 要求处理 JSON 格式错误的异常

# 测试用例：
# read_or_create("data.json")   # 不存在，自动创建并返回 []
# read_or_create("users.json")  # 存在，读取返回内容
# read_or_create("broken.json") # JSON格式错误，打印错误信息

import json
from pathlib import Path

def read_or_create(file_path:str):
    try:
        path = Path(file_path)
        if not path.exists():
            with open(file_path,"w",encoding="utf-8") as f:
                json.dump([],f,ensure_ascii=False,indent=2)
                return []
        else:
            with open(file_path,"r",encoding="utf-8") as f:
                data = json.load(f)
                return data
    except json.JSONDecodeError as e:
        print(f"异常了{e}")

print(read_or_create("config.json"))
```

```python
# 写一个简单的本地通讯录程序
# 数据存储在 contacts.json 文件中
# 实现以下两个函数：
# 1. add_contact(name, phone) 添加联系人，保存到文件
# 2. get_contacts() 读取所有联系人并打印

# 测试用例：
# add_contact("张三", "13800000001")
# add_contact("李四", "13800000002")
# get_contacts()
# 期望输出：
# 张三 - 13800000001
# 李四 - 13800000002
import json
from pathlib import Path
import json
from pathlib import Path

def add_contact(name: str, phone: str):
    file_path = "contacts.json"
    path = Path(file_path)
    if not path.exists():
        with open(file_path, "w", encoding="utf-8") as f:
            json.dump([{"name": name, "phone": phone}], f, ensure_ascii=False, indent=2)
    else:
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)
        # 读完关闭后再写入
        data.append({"name": name, "phone": phone})
        with open(file_path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)

add_contact("张三", "13800000001")
add_contact("李四", "13800000002")

add_contact("张三", "13800000001")
add_contact("李四", "13800000002")
```
