### 1、定义类与实例方法

```python

# 定义一个 Book 类：
# 属性：title（书名）、price（价格）
# 方法：discount(self, percent)，按百分比打折后返回新价格（不修改原 price），比如 percent=20 表示打8折
# 创建一个 Book("Python入门", 100)，调用 discount(20) 并打印结果
# 期望输出：80.0

# 在这里写你的代码

class Book:
    title:str
    price:int

    def __init__(self,title,price):
        self.title = title
        self.price = price

    def discount(self,percent):
        return self.price * (1 -percent/100)

book = Book("Python入门",100)
print(book.discount(20))
```

### 2、类属性做实例计数

```python
# 定义一个 User 类，用一个类属性 count 记录目前一共创建了多少个 User 实例
# 每次 __init__ 调用让 count 加 1
# 创建 3 个 User 实例后，打印 User.count
# 期望输出：3

# 在这里写你的代码

class User:
    count:int = 0

    def __init__(self):
        User.count += 1

u = User()
u2 = User()
u3 = User()
print(User.count)
```

### 3、类方法当替代构造器

```python
# 定义一个 Date 类，属性 year, month, day
# 写一个类方法 from_string(cls, s)，能解析形如 "2024-01-15" 的字符串创建 Date 实例
# 创建实例后打印 f"{year}年{month}月{day}日"
# 调用 Date.from_string("2024-01-15")
# 期望输出：2024年01月15日

# 在这里写你的代码

class Date:
    def __init__(self,year,month,day):
        self.year ,self.month,self.day = year,month,day

    @classmethod
    def from_string(cls,s:str):
        year,month,day = s.split('-')
        return cls(year,month,day)

d = Date.from_string("2024-01-15")
print(f"{d.year}年{d.month}月{d.day}日")
```

### 4、继承与方法重写

```python
# 定义父类 Employee，属性 name, base_salary，方法 salary(self) 返回 base_salary
# 定义子类 Manager(Employee)，多一个属性 bonus，重写 salary(self) 返回 base_salary + bonus
#   （子类 __init__ 要调用 super().__init__() 复用父类逻辑）
# 创建 Manager("Tom", 10000, 5000)，打印 salary()
# 期望输出：15000

# 在这里写你的代码
class Employee:
    name:str
    base_salary:int
    def __init__(self,name,base_salary):
        self.name = name
        self.base_salary = base_salary

    def salary(self):
        return self.base_salary

class Manager(Employee):
    bonus:int
    def __init__(self,name,base_salary,bonus):
        super().__init__(name,base_salary)
        self.bonus = bonus

    def salary(self):
        return self.base_salary + self.bonus

manager = Manager("Tom",10000,5000)
print(manager.salary())
```

### 5、@property 做属性校验

```python
# 定义一个 Account 类，内部用 self._balance 存余额
# 用 @property 暴露只读属性 balance
# 写一个方法 deposit(self, amount)，amount 必须 > 0，否则 raise ValueError("金额必须为正")，校验通过才能给 _balance 加上 amount
# 创建 Account(100)，调用 deposit(50)，打印 balance
# 期望输出：150

# 在这里写你的代码
class Account:
    def __init__(self,balance):
        self._balance = balance

    @property
    def balance(self):
        return self._balance

    def deposit(self,amount):
        if amount <= 0:
            raise ValueError("金额必须为正")
        self._balance += amount

account = Account(100)
account.deposit(50)
print(account.balance)
```

### 6、魔术方法：自定义 Vector

```python
# 定义一个 Vector 类，属性 x, y
# 实现 __str__，打印格式 "Vector(x, y)"，比如 Vector(1, 2) 打印成 "Vector(1, 2)"
# 实现 __add__，支持两个 Vector 相加，返回新的 Vector（对应分量相加）
# 实现 __eq__，比较两个 Vector 的 x, y 是否都相等
# 测试：print(Vector(1, 2) + Vector(3, 4))，并比较 Vector(1,2) == Vector(1,2)
# 期望输出：
# Vector(4, 6)
# True

# 在这里写你的代码

class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"{self.__class__.__name__}({self.x}, {self.y})"

    def __add__(self, other):
        return Vector(self.x + other.x,self.y + other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

print(Vector(1, 2) + Vector(3, 4))
print(Vector(1,2) == Vector(1,2))
```

### 7、多态：统一调用不同类的同名方法

```python
# 定义三个类 Dog、Cat、Duck，都实现一个 speak(self) 方法，分别返回 "汪汪"、"喵喵"、"嘎嘎"
# 创建一个列表装三种动物的实例，用一个循环统一调用 speak() 并打印（不要用 if 判断具体类型）
# 期望输出：
# 汪汪
# 喵喵
# 嘎嘎

# 在这里写你的代码
class Dog:
    def speak(self):
        print("汪汪")

class Cat:
    def speak(self):
        print("喵喵")

class Duck:
    def speak(self):
        print("嘎嘎")

animals = [Dog(),Cat(),Duck()]

for i in animals:
    i.speak()
```

### 8、抽象基类强制约束

```python
# 用 abc 定义一个抽象基类 Payment，包含抽象方法 pay(self, amount)
# 定义子类 AliPay 和 WeChatPay，分别实现 pay，返回 f"支付宝支付了{amount}元" / f"微信支付了{amount}元"
# 创建一个列表装两个子类实例，统一调用 pay(100) 并打印
# 期望输出：
# 支付宝支付了100元
# 微信支付了100元

# 在这里写你的代码

from abc import ABC,abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self,amount):
        pass

class AliPay(Payment):
    def pay(self,amount):
        return  f"支付宝支付了{amount}元"

class WeChatPay(Payment):
    def pay(self,amount):
        return  f"微信支付了{amount}元"

a = AliPay()
w = WeChatPay()
print(a.pay(100))
print(w.pay(100))
```

### 9、用 dataclass 重写一个类

```python
# 用 @dataclass 定义一个 Student 类，字段：name: str, score: int = 0
# 创建 Student("Tom", 90) 和 Student("Tom", 90) 两个实例
# 直接打印第一个实例（验证自动生成的 __repr__），并比较两个实例是否相等（验证自动生成的 __eq__）
# 期望输出：
# Student(name='Tom', score=90)
# True

# 在这里写你的代码
from dataclasses import dataclass

@dataclass
class Student:
    name:str
    score:int

s1 = Student("Tom",90)
s2 = Student("Tom",90)

print(s1)
print(s1 == s2)
```
