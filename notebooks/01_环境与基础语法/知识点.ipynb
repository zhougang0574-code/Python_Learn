# 环境 & 基础语法 · 知识点

### 1、Python 环境与运行方式

```python
# 查看当前 Python 版本和解释器路径
import sys
print(sys.version)
print(sys.executable)

# 两种常见运行方式：
# 1. .py 文件：python main.py（命令行直接跑）
# 2. .ipynb notebook（当前用的）：一个 cell 一个执行单元，可以反复运行、保留中间结果，适合学习和调试

# 虚拟环境（venv）：每个项目独立一套依赖，避免互相污染
# 创建：python -m venv .venv
# 激活（macOS/Linux）：source .venv/bin/activate
# 激活（Windows）：.venv\Scripts\activate
# 类比 Java：相当于每个项目自己的 Maven 本地仓库 + JDK 版本隔离
```

### 2、变量与命名规则

```python
# Python 是动态类型：变量不用声明类型，赋值时类型自动确定，并且可以重新赋值成别的类型
# 对比 Java：Java 是静态类型，String s = "a"; 之后 s 不能变成 int

x = 10
print(type(x))   # <class 'int'>
x = "hello"
print(type(x))   # <class 'str'>，同一个变量可以指向不同类型的对象

# 命名规则：
# - 只能包含字母、数字、下划线，不能以数字开头
# - 约定用 snake_case（如 user_name），不是 Java 常用的 camelCase
# - 常量约定全大写：MAX_SIZE = 100（Python 没有真正的常量，只是命名约定）

# 多重赋值
a, b, c = 1, 2, 3
print(a, b, c)

# 交换变量，不需要中间变量
a, b = b, a
print(a, b)
```

### 3、基本数据类型

```python
# Python 内置基础类型
i = 10            # int，整数，没有 Java 的 int/long 区分，大小不受限
f = 3.14          # float，浮点数，等价于 Java 的 double
s = "hello"       # str，字符串，单引号双引号等价
b = True          # bool，True/False 首字母大写
n = None          # None，等价于 Java 的 null

print(type(i), type(f), type(s), type(b), type(n))

# 用 isinstance() 判断类型，比 type() == 更推荐（支持继承关系判断）
print(isinstance(i, int))
```

### 4、类型转换

```python
# 显式转换：用 int()/float()/str()/bool() 强制转换
print(int("123"))     # 字符串转整数
print(float("3.14"))  # 字符串转浮点数
print(str(123))       # 数字转字符串
print(int(3.99))      # 浮点转整数，直接截断小数部分，不是四舍五入 -> 3

# bool() 的转换规则：0、0.0、""、None、空容器([]/{}/()) 都是 False，其他都是 True
print(bool(0), bool(""), bool([]), bool(1), bool("a"))

# 隐式转换：Python 对不同类型的运算非常严格，不会偷偷转换，要报错
# "1" + 1   # TypeError，str 和 int 不能直接相加，必须显式转换
```

### 5、运算符

```python
# 算术运算符
print(7 / 2)    # 5.5  真除法，结果一定是 float
print(7 // 2)   # 3    地板除，向下取整
print(7 % 2)    # 1    取余
print(2 ** 10)  # 1024 幂运算，对应 Java 的 Math.pow(2, 10)

# 比较运算符：== != > < >= <=，和 Java 一致

# 逻辑运算符：and / or / not（不是 Java 的 && || !）
print(True and False, True or False, not True)

# 赋值运算符：+= -= *= /= 等，和 Java 一致

# 身份运算符 is / is not：比较的是对象是否是同一个（内存地址），不是值是否相等
a = [1, 2]
b = [1, 2]
print(a == b)   # True，值相等
print(a is b)   # False，不是同一个对象

# 成员运算符 in / not in：判断元素是否存在于容器中
print(1 in [1, 2, 3])
print("x" not in "abc")
```

### 6、输入输出

```python
# print 输出，可以一次输出多个值，默认用空格分隔，末尾自动换行
print("a", "b", "c")
print("a", "b", sep="-", end="!\n")   # 自定义分隔符和结尾

# f-string 格式化字符串（推荐写法，对应 Java 的 String.format）
name = "Tom"
age = 18
print(f"{name} is {age} years old")
print(f"{3.14159:.2f}")   # 保留两位小数 -> 3.14

# input() 读取用户输入，返回值永远是 str，需要自己转换类型
# age = int(input("请输入年龄："))
```
