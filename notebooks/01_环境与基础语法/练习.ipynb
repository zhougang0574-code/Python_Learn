### 1、变量与类型转换

```python
# 写一个程序：
# 1. 定义一个字符串变量 price_str = "19.9"
# 2. 把它转换成 float，乘以数量 3，得到总价
# 3. 用 f-string 打印："总价: 59.70 元"（保留两位小数）

# 在这里写你的代码

price_str = "19.9"
total = float(price_str) * 3
print(f"总价：{total:.2f}")
```

### 2、运算符

```python
# 写一个程序，输入两个整数 a 和 b（直接赋值即可，不用 input）
# 分别打印它们的商（真除法）、地板除、余数
# a = 17, b = 5
# 期望输出：
# 商: 3.4
# 地板除: 3
# 余数: 2

# 在这里写你的代码

def func(a:int,b:int):
    print(f"商：{a/b}")
    print(f"地板除：{a//b}")
    print(f"余数：{a%b}")
func(17,5)
```

### 3、身份运算符 vs 比较运算符

```python
# 写一个程序，创建两个内容相同的列表 list1 和 list2
# 分别用 == 和 is 比较，并打印结果，解释为什么结果不同（用注释写出原因）

# 在这里写你的代码

def func():
    list1 = [1,2]
    list2 = [1,2]
    # ==比较的是值是否相等，is比较的是对象地址是否相同，即是否是同一个对象
    print(list1 == list2)
    print(list1 is list2)

func()
```

### 4、字符串格式化

```python
# 写一个程序：
# 定义 name = "Alice", score = 88.5
# 用 f-string 打印："Alice 的成绩是 88.50 分，等级：优秀"
# 规则：score >= 90 优秀以上不用判断，>=80打"优秀"，否则打"及格"（这里只要求拼出字符串，逻辑用 if 写）

# 在这里写你的代码

name = "Alice"
score = 88.5

print(f"{name}的成绩是{score:.2f}分，等级：{'优秀' if score >= 80 else '及格'}")
```

### 5、综合：购物小票

```python
# 写一个程序，模拟打印购物小票：
# 商品名 apple，单价 3.5，数量 4
# 打印格式：
# 商品: apple
# 单价: 3.50 元
# 数量: 4
# 总价: 14.00 元
# 要求：单价和总价都保留两位小数

# 在这里写你的代码

product = 'apple'
sell = 3.5
num = 4
print(f"商品：{product}")
print(f"单价：{sell:.2f}元")
print(f"数量：{num}")
print(f"总价：{sell * num:.2f}元")
```
