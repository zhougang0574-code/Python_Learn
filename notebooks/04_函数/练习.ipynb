### 1、默认参数陷阱

```python
# 写一个函数 add_log(message, logs=None)：
# 把 message 添加到 logs 列表里并返回 logs（注意用 02 章学过的可变默认参数陷阱的正确写法）
# 调用两次 add_log("第一条")、add_log("第二条")，分别打印返回值
# 期望输出：
# ['第一条']
# ['第二条']    注意：不是 ['第一条', '第二条']，说明两次调用互不影响

# 在这里写你的代码

def add_log(message,logs=None):
    if logs is None:
        logs = []
    logs.append(message)
    return logs

print(add_log("第一条"))
print(add_log("第二条"))
```

### 2、*args 求和

```python
# 写一个函数 my_sum(*args)，返回所有传入数字的和
# 调用 my_sum(1, 2, 3, 4, 5)
# 期望输出：15

# 在这里写你的代码

def my_sum(*args):
    return sum(args)

print(my_sum(1, 2, 3, 4, 5))
```

### 3、**kwargs 拼接信息

```python
# 写一个函数 build_info(**kwargs)，把所有关键字参数拼接成 "key=value" 的字符串，用逗号分隔后返回
# 调用 build_info(name="Tom", age=18, city="Beijing")
# 期望输出：name=Tom, age=18, city=Beijing

# 在这里写你的代码

def build_info(**kwargs):
    return ", ".join(f"{k}={v}" for k,v in kwargs.items())

print(build_info(name="Tom", age=18, city="Beijing"))
```

### 4、lambda + sorted 排序

```python
# 有一个列表 students = [{"name": "Tom", "score": 88}, {"name": "Jerry", "score": 95}, {"name": "Alice", "score": 72}]
# 用 lambda 作为 sorted 的 key，按 score 从高到低排序，打印排序后的 name 列表
# 期望输出：['Jerry', 'Tom', 'Alice']

# 在这里写你的代码

students = [{"name": "Tom", "score": 88}, {"name": "Jerry", "score": 95}, {"name": "Alice", "score": 72}]

students_name = [kv["name"]  for kv in sorted(students,key=lambda x: x["score"],reverse=True)]
print(students_name)
```

### 5、global 计数器

```python
# 写一个全局变量 counter = 0
# 写一个函数 click()，每次调用让 counter 加 1（需要用 global 关键字）
# 连续调用 click() 三次，打印 counter
# 期望输出：3

# 在这里写你的代码

counter = 0
def click():
    global counter
    counter += 1

click()
click()
click()
print(counter)
```

### 6、高阶函数

```python
# 写一个函数 apply_twice(func, x)，对 x 应用 func 两次并返回结果
# 例如 apply_twice(lambda x: x * 2, 3) 应该返回 12（3*2=6，6*2=12）
# 用这个函数测试：apply_twice(lambda x: x + 1, 5)
# 期望输出：7

# 在这里写你的代码
def apply_twice(func, x):
    return func(func(x))

v = apply_twice(lambda x: x + 1, 5)
print(v)
```
