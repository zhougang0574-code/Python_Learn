# 数据类型 & 容器 · 知识点

### 1、列表 list

```python
# list 是有序、可变的序列，对应 Java 的 ArrayList，但元素类型可以混着放
nums = [1, 2, 3]
mixed = [1, "a", True, [1, 2]]   # 可以嵌套，可以混类型

# 索引和切片：[start:stop:step]，左闭右开，支持负数索引（-1 表示最后一个）
print(nums[0], nums[-1])
print(nums[0:2])    # [1, 2]
print(nums[::-1])   # [3, 2, 1] 反转

# 常用增删改方法
nums.append(4)        # 末尾追加
nums.insert(0, 0)     # 指定位置插入
nums.extend([5, 6])   # 拼接另一个列表（+= 也行）
nums.remove(3)        # 按值删除第一个匹配项
popped = nums.pop()   # 弹出并返回最后一个元素，可指定下标 pop(0)
print(nums, popped)

nums.sort()            # 原地排序，返回 None（不要写成 nums = nums.sort()）
nums.sort(reverse=True)
print(sorted(nums))    # sorted() 不修改原列表，返回新列表，常用
```

### 2、元组 tuple

```python
# tuple 和 list 很像，但创建后不可变（不能增删改元素）
# 用途：函数一次返回多个值、当字典的 key（list 不行，因为 list 可变不能哈希）
point = (3, 4)
x, y = point   # 解包，对应 Java 里没有的写法，很常用
print(x, y)

# 单元素元组容易踩坑：必须加逗号，不然就是普通括号表达式
not_tuple = (1)      # int，不是 tuple
real_tuple = (1,)    # tuple
print(type(not_tuple), type(real_tuple))

# 函数返回多个值，本质就是返回一个 tuple
def min_max(items):
    return min(items), max(items)

lo, hi = min_max([3, 1, 4, 1, 5])
print(lo, hi)
```

### 3、字典 dict

```python
# dict 是键值对集合，对应 Java 的 HashMap，Python 3.7+ 保证按插入顺序遍历
user = {"name": "Tom", "age": 18}

# 增 / 改：直接赋值，key 不存在就新增，存在就覆盖
user["email"] = "tom@example.com"
user["age"] = 19

# 查：[] 取不存在的 key 会报 KeyError，get() 更安全，可给默认值
print(user["name"])
print(user.get("phone", "未填写"))

# 删
del user["email"]

# 遍历
for k, v in user.items():
    print(k, v)
print(list(user.keys()), list(user.values()))

# setdefault：key 不存在才设置默认值，常用来做"分组累加"
counter = {}
for word in ["a", "b", "a"]:
    counter[word] = counter.get(word, 0) + 1
print(counter)
```

### 4、集合 set

```python
# set 是无序、不重复元素的集合，对应 Java 的 HashSet，最常见用途是去重
nums = [1, 2, 2, 3, 3, 3]
unique = set(nums)
print(unique)   # {1, 2, 3}

a = {1, 2, 3}
b = {2, 3, 4}
print(a | b)    # 并集 union: {1,2,3,4}
print(a & b)    # 交集 intersection: {2,3}
print(a - b)    # 差集 difference: {1}
print(a ^ b)    # 对称差（只在其中一个里出现）: {1,4}

a.add(5)
a.discard(1)    # 删除，元素不存在也不报错（remove() 会报错）
print(a)

# frozenset：不可变版本的 set，可以当字典的 key 或放进另一个 set 里
```

### 5、切片与序列通用操作

```python
# 切片语法对 list / tuple / str 都通用：seq[start:stop:step]
s = "hello world"
print(s[0:5])     # "hello"
print(s[::2])      # 隔一个取一个
print(s[::-1])     # 反转字符串，常用技巧

# 通用操作：len() / in / + / *，list/tuple/str 都支持
print(len(s), "world" in s, [1, 2] + [3, 4], [0] * 3)
```

### 6、推导式（comprehension）

```python
# 列表推导式：用一行代替 for 循环 + append，是 Python 很有代表性的写法
squares = [x * x for x in range(5)]
print(squares)

# 带条件过滤
evens = [x for x in range(10) if x % 2 == 0]
print(evens)

# 等价的传统写法（对比理解）
evens2 = []
for x in range(10):
    if x % 2 == 0:
        evens2.append(x)

# 字典推导式
square_map = {x: x * x for x in range(5)}
print(square_map)

# 集合推导式
unique_lengths = {len(w) for w in ["a", "bb", "ccc", "dd"]}
print(unique_lengths)
```

### 7、可变 vs 不可变 & 浅拷贝/深拷贝

```python
# 不可变类型：int/float/str/bool/tuple —— 一旦创建内容不能改，"修改"实际是创建新对象
# 可变类型：list/dict/set —— 可以原地修改，多个变量名可能指向同一个对象，要小心

a = [1, 2, 3]
b = a            # b 和 a 指向同一个列表对象，不是复制
b.append(4)
print(a)         # [1, 2, 3, 4]，a 也变了！这是最容易踩的坑

# 浅拷贝：复制一层，但嵌套的子对象还是共享的
import copy
c = a.copy()          # 或 list(a) 或 a[:]
c.append(5)
print(a, c)           # a 不受影响，因为顶层是新列表

nested = [[1, 2], [3, 4]]
shallow = nested.copy()
shallow[0].append(99)
print(nested)         # [[1, 2, 99], [3, 4]]，嵌套的子列表还是同一个对象，被改了！

# 深拷贝：连嵌套的子对象也完全复制一份，互不影响
deep = copy.deepcopy(nested)
deep[0].append(100)
print(nested, deep)

# 函数默认参数是可变对象的经典陷阱：默认值只在定义时创建一次，会被多次调用共享
def add_item(item, items=[]):   # 不要这样写！
    items.append(item)
    return items

print(add_item("a"))   # ['a']
print(add_item("b"))   # ['a', 'b'] —— 不是预期的 ['b']，因为 items 复用了上一次的列表
# 正确写法：默认值用 None，函数内部再判断创建新列表
def add_item_fixed(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items
```
