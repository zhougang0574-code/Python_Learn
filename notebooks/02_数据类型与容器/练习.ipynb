### 1、列表操作

```python
# 有一个列表 nums = [5, 3, 8, 1, 9, 3]
# 1. 去掉重复元素（结果不要求有序）
# 2. 把结果从大到小排序
# 3. 打印最大值和最小值
# 期望输出（去重排序后）：[9, 8, 5, 3, 1]，最大值: 9，最小值: 1

# 在这里写你的代码

nums = [5, 3, 8, 1, 9, 3]
v = list(set(nums))
v.sort(reverse=True)
value = sorted(set(nums),reverse=True)
print(value[0])
print(value[-1])
```

### 2、字典统计

```python
# 有一个列表 words = ["apple", "banana", "apple", "orange", "banana", "apple"]
# 统计每个单词出现的次数，存到字典里并打印
# 期望输出：{'apple': 3, 'banana': 2, 'orange': 1}

# 在这里写你的代码
words = ["apple", "banana", "apple", "orange", "banana", "apple"]

d = {}
for w in words:
    if w in d:
        d[w] += 1
    else:
        d[w] = 1

print(d)
```

### 3、集合运算

```python
# 班级 A 的学生：a_class = {"Tom", "Jerry", "Alice"}
# 班级 B 的学生：b_class = {"Alice", "Bob", "Jerry"}
# 打印：两个班都有的学生（交集）、只在 A 班的学生（差集）、所有学生（并集）

# 在这里写你的代码
a_class = {"Tom", "Jerry", "Alice"}
b_class = {"Alice", "Bob", "Jerry"}

print(a_class & b_class)
print(a_class - b_class)
print(a_class | b_class)
```

### 4、元组解包

```python
# 写一个函数 divide(a, b)，同时返回商和余数（用 tuple 返回）
# 调用 divide(17, 5)，用元组解包接收两个返回值并打印
# 期望输出：商: 3, 余数: 2

# 在这里写你的代码

def divide(a,b):
    return (a//b,a%b)

q,r = divide(17,5)
print(f"商: {q}, 余数: {r}")
```

### 5、推导式转换

```python
# 有一个列表 scores = [55, 92, 78, 60, 45, 88]
# 用列表推导式生成一个新列表，只保留及格的分数（>=60）
# 再用字典推导式，把 scores 转换成 {下标: 分数} 的字典（下标从 0 开始）
# 期望输出：
# [92, 78, 60, 88]
# {0: 55, 1: 92, 2: 78, 3: 60, 4: 45, 5: 88}

# 在这里写你的代码

scores = [55, 92, 78, 60, 45, 88]

scores_filter = [ i for i in scores if i >= 60 ]
print(scores_filter)

scores_kv = {k:v for k ,v in enumerate(scores)}
print(scores_kv)
```

### 6、浅拷贝陷阱

```python
import copy

# 有 matrix = [[1, 2], [3, 4]]
# 1. 用 .copy() 浅拷贝一份叫 shallow
# 2. 修改 shallow[0][0] = 99
# 3. 打印 matrix，观察它是否也被改了，并用注释解释原因
# 4. 改用 copy.deepcopy() 重新做一遍，验证 matrix 不会被影响

# 在这里写你的代码

matrix = [[1, 2], [3, 4]]
shallow = matrix.copy()
shallow[0][0] = 99
print(matrix)
#浅拷贝应该只拷贝的内存地址，所以用的是同一份，所以shallow改了，原来的也改了，深拷贝应该是真正的创建了一份新的


deepcopy = copy.deepcopy(matrix)
deepcopy[0][0] = 100
print(deepcopy)
```
